```python
def count_in_line(line):
    forward = ''.join(line).count('XMAS')
    backward = ''.join(line[::-1]).count('XMAS')
    return forward + backward
```

```python
def get_diagonals(data: np.array):
    size = len(data) - 1 
    diagonals = list()
    for i in range(-size, size + 1):
        diagonals.append(data.diagonal(i))
    return diagonals
```

```python
count = 0
for row in data:
	count += count_in_line(row)
for column in data.T:
	count += count_in_line(column)
for diagonal in get_diagonals(data):
	count += count_in_line(diagonal)

data_flipped = np.array([line[::-1] for line in data])
 
for diagonal_t in get_diagonals(data_flipped):
	count += count_in_line(diagonal_t) 
count
```

```python
def get_sub_grids(data):
    subgrids = list()
    for i in range(1, len(data) - 1):
        for j in range(1, len(data) - 1):
            subgrid = data[i-1:i+2, j-1: j+2]
            subgrids.append(subgrid)
    return subgrids
```

```python
get_sub_grids(data)
```

```python
def check_subgrid(subgrid: np.array):
    if subgrid[1][1] != 'A':
        return 0
    if subgrid[0][0] == 'X' or subgrid[0][2] == 'X' or subgrid[2][0] == 'X' or subgrid[2][2] == 'X':
        return 0
    left_diag = (subgrid[0][0] == 'M' and subgrid[2][2] == 'S') or (subgrid[0][0] == 'S' and subgrid[2][2] == 'M')
    right_diag = (subgrid[0][2] == 'M' and subgrid[2][0] == 'S') or (subgrid[0][2] == 'S' and subgrid[2][0] == 'M')
    if left_diag and right_diag:
        return 1
    return 0
        
```

```python
count = 0
for subgrid in get_sub_grids(data):
    count += check_subgrid(subgrid)
count
```
